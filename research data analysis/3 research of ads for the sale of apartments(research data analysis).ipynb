## Предобработка данных

Переименуем столбец близости к центру.

```python
# используем метод rename()
data = data.rename(columns={'cityCenters_nearest' : 'city_centers_nearest'})
# проверим названия столбцов
data.columns
```

### Пропуски

Выведем все пропуски в таблице:

```python
# вызываем функцию isna() совместно с sum()
data.isna().sum()
```

Пропуски присутствуют в 14ти столбцах, есть 49, а есть и 15620. Рассмотрим подробнее.

#### 'ceiling_height'

```python
# вызовем функцию describe()
display(data['ceiling_height'].describe())
# построим гисторгамму методом hist() для более наглядного представления
# задаём размеры графика для наглядности
sb.set(rc={'figure.figsize':(11,8)})
data['ceiling_height'].hist(bins=100,legend=True)
plt.title('Высота потолков')
plt.xlabel("ceiling_height")
plt.ylabel("Частота");
```

Видно, что почти все значения около 3 метров, есть выбросы - максимальное 100, для потолка помещения это явно много. Заполним медианным потолком пропуски, т к при выбросах это значение лучше отражает картину

```python
# функция fillna() находит и заполняет пропуски, передадим ей в аргумент медиану
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
# проверим пропуски
data['ceiling_height'].isna().sum()
```

Значения заполнены

#### 'is_apartment'

```python
# выведем на экран количество значений(value_counts()) и пропусков(isna())
print(data['is_apartment'].value_counts())
print(f'пропусков: {len(data[data["is_apartment"].isna()])}')
```

Предположу, что, скорее всего, люди просто не заполняют эту графу, при создании объявления. Можно заполнить эти занчения параметром 'True', т к квартир ны рынке недвижимости намного больше, чем аппартаментов.  По данным сайта [realty.yandex.ru](https://realty.yandex.ru/) на 25.05.22г. квартир - 54918, аппартаментов - 2908.

```python
# заполняем мелодом fillna(), параметр  True
data['is_apartment'] = data['is_apartment'].fillna(False)
# проверяем пропуски
data['is_apartment'].isna().sum()
```

Столбец 'is_apartment' заполнен.

#### 'floors_total':

```python
# смотрим  количество пропусков
data['floors_total'].isna().sum()
```

```python
# смотрим описание данных методм describe() и строим диаграму на 60 корзин
display(data['floors_total'].describe())
data['floors_total'].hist(bins=60, legend=True)
plt.title('Количество этажей')
plt.xlabel("floors_total")
plt.ylabel("Частота");
```

В данных есть явные пики - 5 и 9 этажей, но есть и выбросы. В Санкт-Петербурге самым высоким зданием является "Лахта-центр", в нём 88 этажей, посмотрим сколько значений находятся после 88:

```python
# функцией len посчитаеи количетво квартир, где этажей больше 88
len(data[data['floors_total'] > 88])
```

Их нет.

## Расчёты и добавление результатов в таблицу

Расчитаем цену квадратного метра:

```python
# добавим цену квадратного метра путём деления цены на площадь
data['square_meter'] = data['last_price'] / data['total_area']
# для наглядности сделаем тип целочисленным и выведем для проверки 5 строк
data['square_meter'] = data['square_meter'].astype(int)
data.head()
```

Приведем в формат 'день недели / месяц / год':

```python
# сделаем новые столбцы для дня, месяца и года использовав функции dt.weekday, dt.month, dt.year
data['day'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year
# напишем функцию, которая возвращает русскиязычный день недели
def rus_day(day):
    if day == 0:
        return 'Понедельник'
    elif day == 1:
        return 'Вторник'
    elif day == 2:
        return 'Среда'
    elif day == 3:
        return 'Четверг'
    elif day == 4:
        return 'Пятница'
    elif day == 5:
        return 'Суббота'
    else:
        return 'Воскресенье'
# напишем функцию, которая возвращает русскиязычный месяц
def rus_month(month):
    if month == 1:
        return 'Январь'
    elif month == 2:
        return 'Февраль'
    elif month == 3:
        return 'Март'
    elif month == 4:
        return 'Апрель'
    elif month == 5:
        return 'Май'
    elif month == 6:
        return 'Июнь'
    elif month == 7:
        return 'Июль'
    elif month == 8:
        return 'Август'
    elif month == 9:
        return 'Сентябрь'
    elif month == 10:
        return 'Октябрь'
    elif month == 11:
        return 'Ноябрь'
    else:
        return 'Декабрь'
# применим функции к соответсвующим столбцам и проверим
data['day'] = data['day'].apply(rus_day)
data['month'] = data['month'].apply(rus_month)
data.head(5)
```

Добавим категоризацию по этажу ('первый/другой/последний'):

```python
# напишем функцию, которая принимает строку и возвращает категорию этажа
# чтобы код не упал из-за пустых значений, заполним константой "-1"
data['floors_total'] = data['floors_total'].fillna(-1)
def flor_category(flors):
    flor = flors['floor']
    floors_total = flors['floors_total']
    if flor == 1:
        return 'Первый'
    elif flor == floors_total:
        return 'Последний'
    elif floors_total == -1:
        return ''
    else:
        return 'Другой'
# создадим новый столбец и зпаишем в него категории этажности, применив функцию flor_category к таблице daya,
# чтобы идти по строкам укажем axis=1   
data['flor_category'] = data.apply(flor_category, axis=1)
# проверим
data.head(5)
```

Внесем в таблицу данные о соотношении жилой площади и кухни к общей площади:

```python
# создадим столбцы с долями жилой площади и кухни, проверим
data['ratio_living'] = round(data['living_area'] / data['total_area'], 2)
data['ratio_kitchen'] = round(data['kitchen_area'] / data['total_area'], 2)
data.head(5)
```
