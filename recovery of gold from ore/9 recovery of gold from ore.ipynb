Данные индексируются датой и временем получения информации (признак date). Соседние по времени параметры часто похожи.
Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков.
Исходный датасет содержит обучающую и тестовую выборки со всеми признаками.

Технологический процесс

Rougher feed — исходное сырье

Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant

    Xanthate — ксантогенат (промотер, или активатор флотации);

    Sulphate — сульфат (на данном производстве сульфид натрия);

    Depressant — депрессант (силикат натрия).

Rougher process (англ. «грубый процесс») — флотация

Rougher tails — отвальные хвосты

Float banks — флотационная установка

Cleaner process — очистка

Rougher Au — черновой концентрат золота

Final Au — финальный концентрат золота

Параметры этапов

air amount — объём воздуха

fluid levels — уровень жидкости

feed size — размер гранул сырья

feed rate — скорость подачи

Наименование признаков

[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:

rougher — флотация

primary_cleaner — первичная очистка

secondary_cleaner — вторичная очистка

final — финальные характеристики

Возможные значения для блока [тип_параметра]:

input — параметры сырья

output — параметры продукта

state — параметры, характеризующие текущее состояние этапа

calculation — расчётные характеристики

### Суммарная концентрация всех веществ на разных стадиях

```python
# считаем концентрации по этапам
rougher_input_sum = geo_full['rougher.input.feed_au']\
            + geo_full['rougher.input.feed_ag']\
            + geo_full['rougher.input.feed_pb']\
            + geo_full['rougher.input.feed_sol']

rougher_output_sum = geo_full['rougher.output.concentrate_au']\
            + geo_full['rougher.output.concentrate_ag']\
            + geo_full['rougher.output.concentrate_pb']\
            + geo_full['rougher.output.concentrate_sol']

primary_cleaner_sum = geo_full['primary_cleaner.output.concentrate_au']\
            + geo_full['primary_cleaner.output.concentrate_ag']\
            + geo_full['primary_cleaner.output.concentrate_pb']\
            + geo_full['primary_cleaner.output.concentrate_sol']

final_sum = geo_full['final.output.concentrate_au']\
            + geo_full['final.output.concentrate_ag']\
            + geo_full['final.output.concentrate_pb']\
            + geo_full['final.output.concentrate_sol']
all_sum = pd.concat([rougher_input_sum, rougher_output_sum, primary_cleaner_sum, final_sum], axis=1)

all_sum = all_sum.rename(columns = {0 : 'rougher_input',
                                    1 : 'rougher_output',
                                    2 : 'primary_cleaner',
                                    3 : 'final_sum'
                                    })
```

Удалим выбросы, где суммарная концентрация ниже 20%

```python
geo_train_full = geo_train.copy()
geo_test_full = geo_test.copy()
geo_full_full = geo_full.copy()

# функци и для подсчета концентраций
def func_del(df):
    df['del_1'] = df['rougher.input.feed_au'] + df['rougher.input.feed_ag'] + df['rougher.input.feed_pb'] + df['rougher.input.feed_sol']
    return 0

def func_del_final(df):
    df['del_1'] = df['rougher.input.feed_au'] + df['rougher.input.feed_ag'] + df['rougher.input.feed_pb'] + df['rougher.input.feed_sol']
    df['del_2'] = df['rougher.output.concentrate_au'] + df['rougher.output.concentrate_ag'] + df['rougher.output.concentrate_pb'] + df['rougher.output.concentrate_sol']
    df['del_3'] = df['primary_cleaner.output.concentrate_au'] + df['primary_cleaner.output.concentrate_ag'] + df['primary_cleaner.output.concentrate_pb'] + df['primary_cleaner.output.concentrate_sol']
    df['del_4'] = df['final.output.concentrate_au'] + df['final.output.concentrate_ag'] + df['final.output.concentrate_pb'] + df['final.output.concentrate_sol']
    return 0

train_shape = geo_train.shape[0]
test_shape = geo_test.shape[0]
full_shape = geo_full.shape[0]

func_del(geo_train)
func_del(geo_test)
func_del_final(geo_full)
# удаление выбросов
geo_train = geo_train[geo_train['del_1'] > 20]
geo_train = geo_train.drop('del_1', axis=1)

geo_test = geo_test[geo_test['del_1'] > 20]
geo_test = geo_test.drop('del_1', axis=1)

geo_full = geo_full[(geo_full['del_1'] > 20) &
                    (geo_full['del_2'] > 20) &
                    (geo_full['del_3'] > 20) &
                    (geo_full['del_4'] > 20)]
geo_full = geo_full.drop(['del_1', 'del_2', 'del_3', 'del_4'], axis=1)

print('из Train удалено', train_shape - geo_train.shape[0], 'строк')
print('из Test удалено', test_shape - geo_test.shape[0], 'строк')
print('из Full удалено', full_shape - geo_full.shape[0], 'строк')
```

**Вывод**

```python
def smape(y_true, y_pred):
    error = abs(y_true - y_pred)
    scale = (abs(y_true) + abs(y_pred)) / 2 + 0.000000001
    result = (error / scale).mean()
    return result
 

def smape_weighted(y_true, y_pred):
    y_true.reset_index(drop=True, inplace=True)
    y_pred = pd.DataFrame(data=y_pred)
    rougher = smape(y_true.iloc[:, 0], y_pred.iloc[:, 0])
    final = smape(y_true.iloc[:, 1], y_pred.iloc[:, 1])
    return 0.25 * rougher + 0.75 * final
```
